```julia
function get_data(m; signal = 5)
    g = CompleteGraph(m);
    s = copy(signal);
    for i = 1:m * 2
        add_vertex!(g);
        add_edge!(g,m+i-1,m+i)
    end
    D = -incidence_matrix(g, oriented = true)';
    D9 = -Matrix(incidence_matrix(PathGraph(9), oriented = true))';
    theta = signal * repeat(pinv(D9) * [s;-s/2;0;s;-s;-s/2;-s/2;s], inner = Int(m/3));
    n = length(theta);
    y = theta + randn(n);
    return Dict([(:y, y), (:theta, theta), (:g, g), (:D, D)])
end
```

```julia
m = 12;
srand(2);
data = get_data(m, signal = 3);
y = data[:y]; theta = data[:theta]; g = data[:g]; D = data[:D];
n = length(y);
R = effective_resistance(g,D)
out = BGSM_effective_resistance(y,D,R, v0 = 1e-1 * 0.2, verbose = false)
ms = model_selection(out);
df = DataFrame(x = repeat(1:n, outer = 3), 
                y = [ms[:t_full];y;theta],
                method = repeat(["BGSM";"data";"true"], inner = length(y)))
Gadfly.plot(df, x = :x, y = :y, color = :method, Geom.line)
```

```julia
function BGSM(y, D, R)
    v0_range = 1e-3 * 1.3.^(0:19); a = Inf; bgsm1 = 0; bgsm2 = 0;
    t = 0;
    for i = 1:length(v0_range)
        tic();
        temp = BGSM_effective_resistance(y,D,R, v0 = v0_range[i], v1 = 1e4, B = n, b = sum(size(D)),
                                        convtol = 1e-8, iter = 100, verbose = false);
        t = t + toc();
        bgsm_ms = model_selection(temp);
        if bgsm_ms[:score] < a
            a = bgsm_ms[:score];
            bgsm1 = temp[:alpha] + temp[:theta];
            bgsm2 = bgsm_ms[:t_full];
        end
    end
    
    return Dict([(:1, bgsm1), (:2, bgsm2), (:t, t)])
end
```

```julia
function sim(m; repnum = 10, signal = 5)
    mse_bgsm = zeros(repnum); mse_fl = zeros(repnum); mse_l0 = zeros(repnum);
    mse_mean = zeros(3); t = zeros(3);
    changes = zeros(3);

    for i = 1:repnum
        d = get_data(m; signal = signal);
        y = d[:y]; theta = d[:theta]; g = d[:g]; D = d[:D]; n = length(y);
        R = effective_resistance(g,D);
        
        tic();
        @rput y;
        R"out =  trendfilter(y, ord=0);";
        t[2] = t[2] + toc();
        R"cv = cv.trendfilter(out);
                 out_fl = coef(out, lambda=cv$lambda.min, verbose = FALSE)$beta;";
        @rget out_fl;
    
        out_bgsm = BGSM(y, D, R);
        t[1] = t[1] + out_bgsm[:t];
        
        tic();
        out_l0 = local_minimizer(y, g, R = R, D = D', lambda = (1 + log(n/10)) * 3, delta = 10/sqrt(n));
        t[3] = t[3] + toc();
        
        changes[1] += sum(abs.(D*out_bgsm[:2]) .> 1e-8);
        changes[2] += sum(abs.(D*out_fl) .> 1e-8);
        changes[3] += sum(abs.(D*out_l0) .> 1e-8);
        
        mse_bgsm[i] = norm(theta - out_bgsm[:2])^2/n;
        mse_fl[i] = norm(theta - out_fl)^2/n;
        mse_l0[i] = norm(theta - out_l0)^2/n;
        
        mse_mean = [mean(mse_bgsm); mean(mse_fl); mean(mse_l0)];
        
    end
    
    return Dict([(:bgsm, mse_bgsm), (:fuzedlasso, mse_fl),
                 (:penl0, mse_l0), (:mean, mse_mean), (:t, t), (:c, changes) ])
end
```
