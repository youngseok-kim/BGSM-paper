```julia
using JuMP, Mosek

m = Model(solver = MosekSolver(QUIET = true))

@variable(m, 0 <= x <= 2)
@variable(m, 0 <= y <= 30)

@objective(m, Max, 5x + 3y)
@constraint(m, 1x + 5y <= 3.0)

#print(m)

status = solve(m)

println("Objective value: ", getobjectivevalue(m))
println("x = ", getvalue(x))
println("y = ", getvalue(y))
```

```julia
function score(A, k = 6; dc = true, sc = [1;2;3])
    ind = find(sum(A,2) .!= 0);
    D = A[ind,:];
    M = 1./sqrt.(sum(D,2));
    X = svds(M .* D, nsv = 6)[1][:U];
    if dc == true
        X = X[:,2:end] ./ X[:,1];
    end
    fig, ax = subplots()
    if length(sc) == 3
        scatter3D(X[:,sc[1]],X[:,sc[2]],X[:,sc[3]])
    elseif length(sc) == 2
        scatter(X[:,sc[1]],X[:,sc[2]])
    else
        error("Error: length(sc) = 2 or 3 for scatter plot ")
    end
    return X
end
```

```julia
score(kushal, 4, dc = true, sc = [1;2])
```
